```python
import scipy.io
from sklearn import svm
import os
import pandas as pd
import numpy as np
from mat4py import loadmat
```

## Reading data from files
### Save y data from one subject to .csv

```python
mydir = 'TrainCharacters'
count = 0
for fname in os.listdir(mydir):
    count+=1
count #number of files
```

```python
y = np.zeros((180,1))
mydir = 'TrainCharacters'
cnt = 0 
for fname in os.listdir(mydir):
    cnt+=1
    fpath = os.path.join(mydir, fname)
    #currentY = scipy.io.loadmat(fpath).get('y')  
    currentY = loadmat(fpath).get('y')
    y = np.vstack((y, currentY))
    if(cnt == 16):
        break
y.shape
```

```python
y = y[180:, :]  #Delete first 180 zero rows
y.shape
```

```python
y
```

```python

# save numpy array as csv file
# save to csv file
np.savetxt('S21_y.csv', y, delimiter=',')
```

### Save X data to csv file

```python
x = np.zeros((180,140))
mydir = 'TrainCharacters'
cnt = 0 
for fname in os.listdir(mydir):
    cnt+=1
    fpath = os.path.join(mydir, fname)
    #currentY = scipy.io.loadmat(fpath).get('y')  
    currentX = loadmat(fpath).get('x')
    #currentY = dataset.get('y')
    x = np.vstack((x, currentX))
    if(cnt == 16):
        break
x.shape
```

```python
x
```

```python
#remove zeros 
x = x[180:, :]
```

```python
x.shape
```

```python
np.savetxt('S21_x.csv', x, delimiter=',')
```

### Read Code variable and save to .csv

```python
code = np.zeros((180,1))
mydir = 'TrainCharacters'
cnt = 0 
for fname in os.listdir(mydir):
    cnt+=1
    fpath = os.path.join(mydir, fname) 
    currentCode = loadmat(fpath).get('code')
    code = np.vstack((code, currentCode))
    if(cnt == 16):
        break
code.shape
```

```python
#remove zeros 
code = code[180:, :]
```

```python
print(code.shape)
code
```

```python
np.savetxt('S21_code.csv', code, delimiter=',')
```
